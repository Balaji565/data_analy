```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress


url = "https://raw.githubusercontent.com/datasets/sea-level-rise/master/data/epa-sea-level.csv"
df = pd.read_csv(url)

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], s=10)


res = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
years_extended = pd.Series(range(df['Year'].min(), 2051))
plt.plot(years_extended,
         res.intercept + res.slope * years_extended,
         'r', label='Fit: all data')

df_2000 = df[df['Year'] >= 2000]
res2 = linregress(df_2000['Year'], df_2000['CSIRO Adjusted Sea Level'])
years_ext2 = pd.Series(range(2000, 2051))
plt.plot(years_ext2,
         res2.intercept + res2.slope * years_ext2,
         'g', label='Fit: 2000 onward')


plt.xlabel("Year")
plt.ylabel("Sea Level (inches)")
plt.title("Rise in Sea Level")
plt.legend()

plt.savefig("sea_level_plot.png")
plt.show()
```
